```python
# Import the necessary modules
from scripts.fetch_message import TOKEN, SECRET
from email_parser.tool import build_gmail_service, read_message_summary, read_message_summary, list_message_ids, fetch_message_json, _body_text
```

```python
# Build a Gmail service object
# This will prompt the user to authorize the application if necessary
# If your token or secrets are at any other location other than the default, specify the paths
gmail = build_gmail_service(token_path=TOKEN, client_secret_path=SECRET)
```

```python
# List message IDs in the INBOX
list_message_ids(gmail, max_results=5, label_ids=["INBOX"])
```

```python
# Print the message summary
# The summary contains the subject, body, and other metadata in a dictionary format
summary = read_message_summary("19a72a0106b60105", gmail_service=gmail)
print(summary)
```

```python
# Fetch the full message JSON
# This includes all headers, parts, and attachments
fetch_message_json(gmail,message_id="19a724d70fc3fe08", format="full")
```

```python
# A for loop to read the top 10 subjects in the INBOX
for message_id in list_message_ids(gmail, max_results=10, label_ids=["INBOX"]):
    summary = read_message_summary(gmail_service=gmail, message_id=message_id)
    print(summary["subject"])
```

```python
# Printing message body 
print(summary["body"])
```
